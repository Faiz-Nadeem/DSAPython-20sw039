```python
import requests

# Fetch a web page
r = requests.get("https://medium.com/@engrfaiznadeem")
print(r.text)
```

```python
import re

# Remove HTML tags using RegEx
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))
```

```python

from bs4 import BeautifulSoup

# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
summaries = soup.find_all("p")
summaries[0]
```

```python
# Extract title
text=soup.find("p").get_text().strip()
```

# Normalization

```python
text = "HEY YOU ARE LOOKING GOOD. It will change your view of the matrix. Are the human people the ones who started the war ? Is AI a bad thing ?"
print(text)
```

```python
text = text.lower()
print(text)
```

```python
#Punctuation Removal
import re

# Remove punctuation characters
text = re.sub(r"[^a-zA-Z0-9]", " ", text)
print(text)
```

```python
words = text.split()
print(words)
```
